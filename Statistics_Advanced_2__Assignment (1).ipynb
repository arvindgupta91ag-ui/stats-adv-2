1. What is hypothesis testing in statistics?

- Hypothesis testing is a statistical method used to evaluate assumptions about a population parameter using sample data. It involves formulating a null hypothesis (H₀) and an alternative hypothesis (H₁), then using test statistics and p-values to decide whether to reject H₀. It is critical for making data-driven decisions and validating claims.

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

- The significance level (α) is the threshold for rejecting H₀, commonly set at 0.05. It represents the probability of making a Type I error (rejecting H₀ when it is true). If the p-value ≤ α, we reject H₀; otherwise, we fail to reject it.

4. What are Type I and Type II errors? Give examples of each.


- Type I Error: Rejecting H₀ when it is true. Example: Concluding a medicine works when it actually doesn’t.

Type II Error: Failing to reject H₀ when H₁ is true. Example: Concluding a medicine doesn’t work when it actually does

5.  What is the difference between a Z-test and a T-test? Explain when to use
each.

Z-test: Used when population variance is known and sample size is large (n ≥ 30).

T-test: Used when population variance is unknown and/or sample size is small (n < 30).
Both test means, but the t-test accounts for extra uncertainty in small samples.

6.  Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate binomial distribution (n=10, p=0.5)
data = np.random.binomial(n=10, p=0.5, size=1000)

# Plot histogram
plt.hist(data, bins=11, edgecolor='black', color='skyblue')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

7.  Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)

```python
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Hypothesized mean (μ0 = 50)
z_stat = (mean - 50) / (std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

8.  Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and CI
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

# Plot histogram
plt.hist(data, bins=30, color='lightgreen', edgecolor='black')
plt.title("Normal Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

9.  Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
data = np.array([10, 12, 13, 15, 18, 20, 22, 25, 30])

# Calculate Z-scores
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

print("Z-scores:", z_scores)

# Plot histogram of standardized data
plt.hist(z_scores, bins=5, edgecolor='black', color='orange')
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
